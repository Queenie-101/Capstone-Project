```python
#Library imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data
df=pd.read_csv('/content/penguins_size.csv')
df.head()
```

```python
#Information about the data
df.info()
```

```python
#Check for null values
df.isnull().sum()
```

```python
#Drop the null values
df.dropna(inplace=True)
```

```python
#Get the general relationship between variables
sns.pairplot(data=df,hue='species',palette='cubehelix')
```

```python
#Count of gender in the dataset
df['sex'].value_counts()
```

```python
#Plot of male against female
sns.countplot(data=df,x='sex',palette='coolwarm')
plt.show()
```

#Findings
* There is no major difference in the number of male and female penguins.

```python
#Plot the number of penguins in islands
sns.countplot(data=df,x='island',palette='summer')
plt.show()
```

#Findings
* Biscoe Island has the largest number of penguins.
* Torgersen Island has the least number of penguins.

```python
#Compare the genders in different islands
sns.countplot(data=df,x='island',hue='sex',palette='RdPu')
```

#Findings
* Torgersen Island is the only island with more female penguins than make penguins.

```python
#Species count
sns.countplot(data=df,x='species',palette='rainbow')
```

#Findings
* The Adelie species is the largest species.
* The Chinstrap species, is an endangered species.

```python
#Count of each species
df['species'].value_counts()
```

```python
#Species against gender
sns.countplot(data=df,x='species',hue='sex',palette='rocket')
```

#Findings
* There is no major difference between male and female counts in Adelie and Chinstrap species.
* In Gentpp species, the number of male penguins is slightly higher than the number of female penguins.

```python
#Species vs Island
sns.countplot(data=df,x='species',hue='island',palette='magma')
```
